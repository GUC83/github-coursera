## FIRST PART : GETTING THE DATA

```python
link = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

df = pd.read_html(link)[0]
df
```

```python
#discarding the Not Assigned
df = df[df['Borough'] != 'Not assigned']
df.head()
```

```python
#SSeparating with commas and eliminating duplicates
df['Neighbourhood'] = df.groupby("Postcode")["Neighbourhood"].transform(lambda neigh: ', '.join(neigh))
df.drop_duplicates(subset ="Postcode", inplace = True)
df.head(20)
```

```python
df.shape
```

```python
filename = 'https://cocl.us/Geospatial_data'

geodata = pd.read_csv(filename)
geodata.head(20)
```

```python
df.index = range(df.shape[0])
finaldf = df.merge(geodata, left_on = 'Postcode', right_on = 'Postal Code', how = 'left').drop (column= 'Postal Code')
finaldf
```
